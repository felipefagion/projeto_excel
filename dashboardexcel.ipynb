# Conclusão 

Em conclusão, a análise dos dados revelou muitas coisas importante sobre o desempenho das marcas e subcategorias na categoria de higiene bucal. A marca A obteve um aumento significativo nas unidades vendidas ao optar por não aumentar seu preço na subcategoria mais relevante, enquanto a marca B conseguiu aumentar seus valores de vendas ao aumentar a margem de preço, especialmente na subcategoria de creme dental.
